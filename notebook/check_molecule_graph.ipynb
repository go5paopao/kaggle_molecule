```python
data_st = st[st.molecule_name == data[0]]
data_st = data_st.merge(data_st, on="molecule_name", how="left",suffixes=("_0","_1"))
data_st = data_st[data_st.atom_index_0 != data_st.atom_index_1].reset_index(drop=True)
data_st["dist"] = np.linalg.norm(data_st[["x_0","y_0","z_0"]].values - data_st[["x_1","y_1","z_1"]].values, axis=1)
```
